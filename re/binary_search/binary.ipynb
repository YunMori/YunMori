```python
#BOJ_12015 05/17 30분 31초
import sys, bisect
# sys.stdin = open('/Users/yunjongseo/Downloads/input (10).txt', 'rt')
N = int(sys.stdin.readline())
A = list(map(int, sys.stdin.readline().split()))
stack = []
cnt = 0
for i in range(N):
    x = bisect.bisect_left(stack, A[i])
    if x == len(stack):
        stack.append(A[i])
        cnt += 1
    else:
        stack[x] = A[i]
print(cnt)
```
