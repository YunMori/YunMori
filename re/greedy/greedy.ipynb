```python
#1049 Oct 19, 2025
import sys
import math

# sys.stdin = open('/Users/yunjongseo/Downloads/sample_input (2).txt', 'rt')

#N = 끊어진 기타줄 갯수
#M = 기타줄 브랜드 갯수
N, M = map(int, sys.stdin.readline().split())

min_a = 1001
min_b = 1001
for _ in range(M):
    a, b = map(int, sys.stdin.readline().split())
    min_a = min(a, min_a)
    min_b = min(b, min_b)

op_1 = min_a * math.ceil(N/6)
op_2 = min_b*N
op_3 = min_a * (N//6) + min_b*(N%6)
print(min(op_1, op_2, op_3))
```

```python
#1080 Oct 22, 2025
import sys
# sys.stdin = open('/Users/yunjongseo/Downloads/sample_input (2).txt', 'rt')

def flip(matrix, x, y):
    for i in range(x, x+3):
        for j in range(y, y+3):
            matrix[i][j] = 1 - matrix[i][j]
            
            
N, M = map(int, sys.stdin.readline().split())
A = list(list(map(int, sys.stdin.readline().strip())) for _ in range(N))
B = list(list(map(int, sys.stdin.readline().strip())) for _ in range(N))
cnt = 0

for i in range(N-2):
    for j in range(M-2):
        if A[i][j] != B[i][j]:
            flip(A, i, j)
            cnt += 1
if A == B:
    print(cnt)
else:
    print(-1)
```

```python
#1138 Oct 25, 2025
import sys
# sys.stdin = open('/Users/yunjongseo/Downloads/sample_input (2).txt', 'rt')

N = int(sys.stdin.readline())
M = list(map(int, sys.stdin.readline().split()))

ml= [0]*N

for i in range(N):
    cnt = M[i]
    for j in range(N):
        if cnt == 0 and ml[j] == 0:
            ml[j] = i+1
            break
        if ml[j] == 0:
            cnt -= 1
print(' '.join(map(str, ml)))
        
```

```python
#1141 Oct 25, 2025
import sys
# sys.stdin = open('/Users/yunjongseo/Downloads/sample_input (2).txt', 'rt')
N = int(sys.stdin.readline())
words = sorted(list(sys.stdin.readline().strip() for _ in range(N)), key= lambda x: -len(x))
answer = 0
for i in range(N):
    check = False
    for j in range(i):
        if words[j].startswith(words[i]):
            check = True
            break
    if not check:
        answer += 1
print(answer)
```
