```python
#12919 Nov 19, 2025

import sys
from collections import deque

# sys.stdin = open('/Users/yunjongseo/Downloads/sample_input (3).txt', 'rt')

S = sys.stdin.readline().strip()
T = sys.stdin.readline().strip()
S_list = list(i for i in S)
T_list = list(i for i in T)

Q = deque()
Q.append(T_list)

while Q:
    a = Q.popleft()
    if a == S_list:
        print(1)
        break
    elif len(a) < len(S_list):
        print(0)
        break
    
    b = a[:]
    c = a[:]
    
    if b[-1] == 'A':
        b.pop()
        Q.append(b)
        
    if c[0] == 'B':
        c.reverse()
        c.pop()
        Q.append(c)
else:
    print(0)
```
