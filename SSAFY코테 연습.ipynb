```python
#SW expert 1209
def Sum(table):
    Sum = []
    #열
    for y in range(100):
        Sum_d = []
        for x in range(100):
            Sum_d.append(table[y][x])
        Sum.append(sum(Sum_d))
    #행
    for x in range(100):
        Sum_d = []
        for y in range(100):
            Sum_d.append(table[y][x])
        Sum.append(sum(Sum_d))
    #대각선
    Sum_d = []
    for i in range(100):
        Sum_d.append(table[i][i])
    Sum.append(sum(Sum_d))
    #역대각선
    Sum_d = []
    for i in range(100):
        Sum_d.append(table[i][99-i])
    Sum.append(sum(Sum_d))

    answer = max(Sum)
    return answer

for test_case in range(1, 11):
    N = int(input())
    table = []
    for _ in range(100):
        table.append(list(map(int,input().split())))
    answer = Sum(table)
    print(f'#{N} {answer}')
```

```python
#SW expert 2806

def safe_checker(queen, row, col):
    for r in range(row):
        c = queen[r]
        if c == col or abs(r - row) == abs(c - col):
            return False
    return True
        
    

def solution(queen, row, col):
    if row == N:
        return 1
    cnt = 0
    for col in range(N):
        if safe_checker(queen, row, col):
            queen[row] = col
            cnt += solution(queen, row+1, N)
    return cnt

T = int(input())
for teat_case in range(1, T+1):
    N = int(input())
    queen = [-1]*N
    result = solution(queen, 0, N)
    print(f'#{teat_case} {result}')
            
```

```python
#SW expert 1959

def solution_A(N, M, A, B, V = []):
    if len(A) > len(B):
        return max(V)
    result = 0
    for i in range(N):
        result += (A[i]*B[i])
    B.popleft()
    V.append(result)
    solution_A(N, M, A, B, V)
    return max(V)

def solution_B(N, M, A, B, V = []):
    if len(A) < len(B):
        return max(V)

    result = 0
    for i in range(M):
        result += (A[i]*B[i])
    A.popleft()
    V.append(result)
    solution_B(N, M, A, B, V)
    return max(V)
        

# import sys
from collections import deque
# sys.stdin = open('/Users/yunjongseo/Downloads/input.txt', 'rt')
T = int(input())
for test_case in range(1, T+1):
    N, M = map(int, input().split())
    A = deque(map(int, input().split()))
    B = deque(map(int, input().split()))
    if N <= M:
        result = solution_A(N, M, A, B, V=[])
    else:
        result = solution_B(N, M, A, B, V=[])
    print(f'#{test_case} {result}')
```
