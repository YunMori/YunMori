```python
#SW expert 1244 다시 공부
def dfs(NL, C, memo):
    # 현재 숫자판 상태를 문자열로 만듦
    current_num = ''.join(map(str, NL))
    
    # 메모이제이션: 현재 상태와 남은 교환 횟수가 이미 저장되어 있다면 그 값 반환
    if (current_num, C) in memo:
        return memo[(current_num, C)]
    
    # 교환 횟수를 모두 사용한 경우 현재 숫자판의 정수값 반환
    if C == 0:
        return int(current_num)

    max_value = 0
    V = len(NL)
    
    # 가능한 모든 자리 교환
    for a in range(V):
        for b in range(a + 1, V):
            # 자리 교환
            NL[a], NL[b] = NL[b], NL[a]
            
            # 재귀 호출하여 최대값 갱신
            max_value = max(max_value, dfs(NL, C - 1, memo))
            
            # 자리 복원
            NL[b], NL[a] = NL[a], NL[b]

    # 메모이제이션에 결과 저장
    memo[(current_num, C)] = max_value
    return max_value

# 테스트 케이스 실행
T = int(input())
for test_case in range(1, T + 1):
    N, C = map(int, input().split())
    NL = list(map(int, str(N)))  # 정수를 리스트로 변환
    memo = {}  # 메모이제이션 딕셔너리
    answer = dfs(NL, C, memo)
    print(f'#{test_case} {answer}')
    
    
'''
def solution(NL,C,Max_value):
    V = len(NL)
    if C == 0:
        return Max_value
    
    for a in range(V):
        for b in range(a+1,V):
            NL[a],NL[b]=NL[b],NL[a]
            R = int(''.join(map(str,NL)))
            Max_value = max(Max_value, solution(NL,C-1,R))
            NL[b],NL[a] = NL[a],NL[b]
    return Max_value

T = int(input())
for test_case in range(1,T+1):
    N,C = map(int,input().split())
    NL = list(map(int,str(N)))
    Max_value = 0
    answer = solution(NL,C,Max_value)
    print(f'#{test_case} {answer}')
'''
```

```python
#SW expert 2001
def mapping(List,N, M, answer = 0):
    for y in range(N-M+1):
        for x in range(N-M+1):
            a = []#왜 여기서 초기화 해야하지?
            for ny in range(M):
                for nx in range(M):
                    a.append(List[y+ny][x+nx])
            b = sum(a)
            if answer<b:
                answer = b
    return answer
                    

T = int(input())
for test_case in range(1, T + 1):
    List = []
    N, M = map(int, input().split())
    for _ in range(N):
        O = list(map(int, input().split()))
        List.append(O)
    V = mapping(List, N, M)
    print(f'#{test_case} {V}')
```

```python
#SW expert 1220

def Magnetic (table, N):
    for x in range(N):
        for y in range(N-1):
            if table[y][x] == 1 and table[y+1][x] == 0:
                table[y+1][x] = table[y][x]
                table[y][x] = 0
        for y in range(N, -1):
            if table[y][x] == 2 and table[y+1][x] == 0:
                table[y+1][x] = table[y][x]
                table[y][x] = 0
    count = 0
    for x in range(N):
        for y in range(N-1):
            if table[y][x] == 1 and table[y+1][x] == 2:
                count += 1
    return count

for test_case in range(1,11):
    N = int(input())
    table = []
    for _ in range(N):
        table.append(list(map(int,input().split())))
    answer = Magnetic(table, N)
    print(f'#{test_case} {answer}')
```

```python
#SW expert 1209
def Sum(table):
    Sum = []
    #열
    for y in range(100):
        Sum_d = []
        for x in range(100):
            Sum_d.append(table[y][x])
        Sum.append(sum(Sum_d))
    #행
    for x in range(100):
        Sum_d = []
        for y in range(100):
            Sum_d.append(table[y][x])
        Sum.append(sum(Sum_d))
    #대각선
    Sum_d = []
    for i in range(100):
        Sum_d.append(table[i][i])
    Sum.append(sum(Sum_d))
    #역대각선
    Sum_d = []
    for i in range(100):
        Sum_d.append(table[i][99-i])
    Sum.append(sum(Sum_d))

    answer = max(Sum)
    return answer

for test_case in range(1, 11):
    N = int(input())
    table = []
    for _ in range(100):
        table.append(list(map(int,input().split())))
    answer = Sum(table)
    print(f'#{N} {answer}')
```

```python
#SW expert 1234
def solution(len , N , i = 0):
    if i == len:
        return N
    if N[i] != N[i+1]:
       i += 1
       solution(len, N, i)
    else:
        N.pop(N.index(N[i]))
        N.pop(N.index(N[i+1]))
        i+=1
        solution(len, N, i)
    return N
        
for test_case in range(1, 11):
    len, N = map(int, input().split())
    N = list(map(int,str(N)))
    len = len-1
    answer = solution(len, N)
    print(f'#{test_case} {answer}')
```

```python
#SW expert 2806

def safe_checker(queen, row, col):
    for r in range(row):
        c = queen[r]
        if c == col or abs(r - row) == abs(c - col):
            return False
    return True
        
    

def solution(queen, row, col):
    if row == N:
        return 1
    cnt = 0
    for col in range(N):
        if safe_checker(queen, row, col):
            queen[row] = col
            cnt += solution(queen, row+1, N)
    return cnt

T = int(input())
for teat_case in range(1, T+1):
    N = int(input())
    queen = [-1]*N
    result = solution(queen, 0, N)
    print(f'#{teat_case} {result}')
            
```
